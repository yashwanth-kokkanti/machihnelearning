```
### Sentiment Analysis 

''' Sentiment Analysis is one of the applications of NLP. It is used to extract subjective information such as opinions or attitudes from text, speech or tweets.`
    In simple terms sentiment analysis involves classfying/identifying text to positive, negative or neutral. '''

  
### Problem Statement

''' The objective of this task is to detect hate/negative speech in tweets. If there is negative word like racist or sexist or any of the associative workds, then 
    it is considered as a negative sentiment. 

    In the Data Label 1 represents that it is a racist/negative statement and 0 represents positive statement. ''' 

### 
''' Evaluation Metric is given as F1-Score.  = 2 * precision * recall / (precision + recall)
     - Accuracy can be highly influenced by True negatives 
     - F1 Score balances between precision and recall even when there is an uneven class distribution. '''
```

```
## Data Cleaning - Clean Text data, remove stop words using nltk . 

## Let us combine train and test data. 

combi = train.append(test, ignore_index=True)
```

```
combi.shape
```

```
## Write a simple function to remove unwanted Patterns 

def remove_pattern(input_txt, pattern):
  ''' 
    remove unwanted patters 
  '''
  var = re.findall(pattern, input_txt)
  for i in var:
    input_txt = re.sub(i , '', input_txt)
  return input_txt


## Following text will be removed .
## @user is removed due to privacy. 
## Punctuations, numbers, special characters, 
```

```
## Remove twitter handles like @user 

combi['tidy_tweet'] = np.vectorize(remove_pattern)(combi['tweet'], '@[\w]*')  ## removing all words atarting with '@'
combi.head()
```

```
## Removing Puncutations 

combi['tidy_tweet'] = combi['tidy_tweet'].str.replace('[^a-zA-Z#]', " ")
combi.head()
```

```
## Removing Short words like his, her, the, etc, not. 

combi['tidy_tweet']  = combi['tidy_tweet'].apply(lambda x: ' '.join( [w for w in x.split() if len(w) > 3] ) )
combi.head()
```

```
## Text Normalization . We will use PorterStemmer to normalize (to convert to root word)
## before that we need to tokanize sentences into words 

tokenized_tweet = combi['tidy_tweet'].apply(lambda x: x.split()) ## Tokenizing , just splitting with space 
tokenized_tweet.head()
```

```
from nltk.stem.porter import *

stemmer = PorterStemmer()
tokenized_tweet = tokenized_tweet.apply(lambda x: [stemmer.stem(i) for i in x] ) # Stemming

tokenized_tweet
```

```
## Now lets attach these back together and this can be done using nltk MosesDEtokenizer function 

for i in range (len(tokenized_tweet)):
  tokenized_tweet[i] = ' '.join(tokenized_tweet[i])
combi['tidy_tweet'] = tokenized_tweet

```

```
## few words with hashtag words has more weightage to the sentences few words like #faminism, #psycho are examples.

## now let us write a function to store trend_terms in two lists. 
## one for storing non-racist words and other for storing racist words

def hashtag_extract(x):
  hashtags = []
  for i in x:
    ht = re.findall(r'#(\w+)', i)
    hashtags.append(ht)
  return hashtags

## Extarcting hashtags from non racist tweets 

HT_regular = hashtag_extract(combi['tidy_tweet'][combi['label'] == 0])

## Extracting hashtags from racist tweets 

HT_negative = hashtag_extract(combi['tidy_tweet'][combi['label'] == 1])

## unnesting list .. Variable of nested list. Example given below 

# >>> ab = [['hi', 'hello']]
# >>> ab = sum(ab, [])
# >>> ab
# ['hi', 'hello']

HT_regular = sum(HT_regular, [])
HT_negative = sum(HT_negative, [])

```

```
## Let us plot top words for non racist words 

a = nltk.FreqDist(HT_regular)
d = pd.DataFrame({'Hashtag': list(a.keys()), 'Count': list(a.values())})

## selecting top 20 Rows 
d = d.nlargest(columns='Count', n=20)
plt.figure(figsize=(16, 5))
ax = sns.barplot(data=d, x='Hashtag', y='Count')
ax.set(ylabel='Count')
plt.show()
```

```
## Let us plot for Racist Tweets 


b = nltk.FreqDist(HT_negative)
e = pd.DataFrame({'Hashtag': list(b.keys()), 'Count': list(b.values())})

## selecting top 20 Rows 
e = e.nlargest(columns='Count', n=20)
plt.figure(figsize=(16, 5))
ax = sns.barplot(data=e, x='Hashtag', y='Count')
ax.set(ylabel='Count')
plt.show()
```
