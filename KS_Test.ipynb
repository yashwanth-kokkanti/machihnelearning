```python
# KS test is used to compare two plots (another way of comparing two distribution)
#Imports 

import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt

# generate a random variable 
x = stats.norm.rvs(size=1000)
## print (x) prints 1000 random variables
sns.set_style("whitegrid")
sns.kdeplot(np.array(x), bw=1) # kernel density estimated plot
plt.show()
```

```python
stats.kstest(x, 'norm')  ## since the p Value is far more than statistic two plots are randomly distributed.
```

```python
# checking with Uniform distribution
y = np.random.uniform(1, 5, 1000)
sns.kdeplot(np.array(y), bw=0.1)
plt.show()
```

```python
stats.kstest(y, 'norm')
```
