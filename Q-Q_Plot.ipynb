```python
# Example of Q-Q Plot Using scipy

# Import statements 
import numpy as np
import pylab
import scipy.stats as stats
```

```python
# Generate Normal/Gaussian distribution
std_normal = np.random.normal(loc=0, scale=1, size=1000)
# Print 0 - 100 percentile of given distribution of 1000 values
for i in range (0, 101, 25):
    print (i, np.percentile(std_normal, i))
```

```python
# for sake of visualising values printing all 100 percentiles
for i in range(1, 101):
    print (i, np.percentile(std_normal, i))
```

```python
# generate samples of gaussian dist with loc = 10 and scale = 2 
measurements = np.random.normal(loc=10, scale=2, size=500)

# comparing two distributions using Q-Q plot (Probplot in scipy)
stats.probplot(measurements, dist='norm', plot=pylab)
pylab.show()
```

```python
# try with different distribution (uniform)
measurements_1 = np.random.uniform(low=-1, high=1, size=10000)

stats.probplot(measurements_1,  dist='norm', plot=pylab)
pylab.show()
```
